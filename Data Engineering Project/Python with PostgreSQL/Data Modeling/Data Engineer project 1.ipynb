# Create database and table for these 3 Datasets

```python
cur, conn = create_database() 
```

```python
sales_table_create = (""" CREATE TABLE IF NOT EXISTS salesData(
Order_ID int, 	
Product	varchar,
Purchase_Address varchar,	
Quantity_Ordered int,	
Price_Each varchar
)""")
```

```python
cur.execute(sales_table_create)
conn.commit()
```

```python
amazon_table_create = (""" CREATE TABLE IF NOT EXISTS amazon(
product_id varchar,
product_name varchar,
discounted_price varchar,	
actual_price varchar
)""")
```

```python
cur.execute(amazon_table_create)
conn.commit()
```

```python
supermarket_table_create = (""" CREATE TABLE IF NOT EXISTS supermarket(
Branch varchar,
City varchar,
Gender varchar,
Quantity int,
Payment varchar
)""") 
cur.execute(supermarket_table_create)
conn.commit()
```
